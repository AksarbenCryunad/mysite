```python
import pandas as pd
import invest
```

```python
# 주석 코드와 시작시간을 이용하여 데이터를 로드
input_code = 'AAPL'
input_time = '2020-01-01'
```

```python
# 데이터 로드
df = invest.load_data(input_code, input_time)
```

```python
df.head()
```

```python
# Invest라는 class 생성
invest_class = invest.Invest(
    df, _col = 'Close', _start = input_time
)
```

```python
# 유저가 보낸 데이터 중 투자 방식
input_type = 'boll'
```

```python
if input_type == 'bhn':
    result = invest_class.buyandhold()
elif input_type == 'boll':
    result = invest_class.bollinger()
elif input_type == 'mmt':
    result = invest_class.momentum()
```

```python
result.tail()
```

```python
result.reset_index(inplace=True)
result.head()
```
